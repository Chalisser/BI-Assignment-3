```python
def now() -> str:
    """
    Returns the current time in ISO 8601 format with UTC timezone in the following format:
    YYYY-MM-DDTHH:MM:SS.sssZ
    """
    timestamp = datetime.datetime.now(datetime.timezone.utc)
    timestamp_formated = timestamp.strftime("%Y-%m-%dT%H:%M:%S.%f")[:-3]  +"Z"

    return timestamp_formated
```

**What not do do**

Do not use [blank nodes](https://www.w3.org/wiki/BlankNodes).

PROV-O uses blank nodes to connect multiple elements with each other.
Such blank nodes (such as _:association) should not be used.
Instead, assign a fixed node ID such as
:5119fcd7-b571-41e0-9464-a37c7be0f574 by generating them outside of the
notebook.
We suggest that, for each setting where such a blank node is needed to
connect multiple elements, you create a unique hash (using uuid.uuid4())
and keep this as hard-coded identifier for the blank node. The template
notebook contains examples of this. Do *not* use these provided values,
as otherwise, your provenance documentations will all be connected via
these identifiers!
Also, do not generate them dynamically in every cell execution, e.g. by
using uuid.uuid4() in a cell. This would generate many new linking nodes
for connecting the same elements.
Compute one for each node (cell) where you need them and make sure to
use the same one on each re-execution of the notebook.

```python
check_outliers_code_writer = student_a

def check_outliers(
    data: pd.DataFrame,
    threshold: float = 3.0,
    columns=(
        'Age at enrollment',
        'Admission grade',
        'Previous qualification (grade)',
        'Curricular units 1st sem (grade)',
        'Curricular units 2nd sem (grade)',
        'Unemployment rate',
        'Inflation rate',
        'GDP',
    ),
    max_examples_per_col: int = 5
) -> dict:
    """
    Identify potential outliers via z-score on selected continuous columns.
    Returns a compact report (count + up to N example indices/z-scores per column).
    """
    report = {}
    tmp = data.reset_index(drop=True)

    for col in columns:
        if col not in tmp.columns:
            # Column not available -> skip
            continue

        values = pd.to_numeric(tmp[col], errors="coerce")
        mean = values.mean()
        std = values.std()

        if std == 0 or np.isnan(std):
            report[col] = {"n_outliers": 0, "examples": []}
            continue

        z_scores = (values - mean) / std
        mask = np.abs(z_scores) > threshold

        # collect outliers
        outlier_indices = values[mask].index.tolist()

        # sort examples by absolute z-score (largest first)
        outlier_indices_sorted = sorted(outlier_indices, key=lambda i: abs(z_scores.loc[i]), reverse=True)

        examples = [
            {"index": int(i), "z_score": float(z_scores.loc[i])}
            for i in outlier_indices_sorted[:max_examples_per_col]
        ]

        report[col] = {
            "n_outliers": int(len(outlier_indices)),
            "threshold": float(threshold),
            "examples": examples
        }

    return report


start_time_co = now()
outliers_report = check_outliers(data, threshold=3.0)
end_time_co = now()

print(json.dumps(outliers_report, indent=2))
#############################################
# Documentation
#############################################
# Three steps in this process:
# 1) an activity creates an outcome (here: an outlier report)
# 2) an activity inspects the outcome and derives a decision
# 3) a follow-up activity changes the data (will be done in Data Preparation, if needed)

#############################################
# 1) Activity: Checking for outliers and creating the report
#############################################

co_ass_uuid_executor = "15085e9d-15f1-4727-9b6e-776dd07fcd08"  # generate once
check_outliers_executor = [
    f':check_outliers prov:qualifiedAssociation :{co_ass_uuid_executor} .',
    f':{co_ass_uuid_executor} prov:agent :{executed_by} .',
    f':{co_ass_uuid_executor} rdf:type prov:Association .',
    f':{co_ass_uuid_executor} prov:hadRole :{code_executor_role} .',
]
engine.insert(check_outliers_executor, prefixes=prefixes)

co_ass_uuid_writer = "cd4970df-9f40-4bb1-8fad-e4dc4fcdd284"  # generate once
co_comment = """
Potential outliers were identified using a z-score approach (|z| > 3.0) on selected continuous variables
(e.g., age, grades, macro-economic indicators). This is an exploratory analysis in the Data Understanding phase.
No data is modified in this step.
"""

check_outliers_activity = [
    ':check_outliers rdf:type prov:Activity .',
    ':check_outliers sc:isPartOf :data_understanding_phase .',
    ':check_outliers rdfs:comment \'Data Understanding\' .',
    f':check_outliers rdfs:comment """{co_comment}""" .',
    f':check_outliers prov:startedAtTime "{start_time_co}"^^xsd:dateTime .',
    f':check_outliers prov:endedAtTime "{end_time_co}"^^xsd:dateTime .',

    f':check_outliers prov:qualifiedAssociation :{co_ass_uuid_writer} .',
    f':{co_ass_uuid_writer} prov:agent :{check_outliers_code_writer} .',
    f':{co_ass_uuid_writer} rdf:type prov:Association .',
    f':{co_ass_uuid_writer} prov:hadRole :{code_writer_role} .',

    ':check_outliers prov:used :data .',

    ':outlier_report rdf:type prov:Entity .',
    f':outlier_report rdfs:comment """{json.dumps(outliers_report, indent=2)}""" .',
    ':outlier_report prov:wasGeneratedBy :check_outliers .',
]
engine.insert(check_outliers_activity, prefixes=prefixes)

#############################################
# 2) Activity: Inspecting the report and deriving a decision
#############################################

start_time_ior = now()

ior_ass_uuid_executor = "6eaa2c0a-e592-4d85-b37f-d695844910cf"  # generate once
ior_comment = """
The outlier report was inspected. Outliers mainly occur in 'Age at enrollment' and in grade-related variables,
while no outliers were detected for macro-economic variables at z>3.0.

Interpretation: These are plausible rare cases (e.g., older students or extreme grades) and not necessarily data errors.
Decision: Do not remove observations in Data Understanding. If required, consider robust preprocessing (e.g., robust scaling
or winsorization) in the Data Preparation phase.
"""

inspect_outlier_report_executor = student_a

inspect_outlier_report_activity = [
    ':inspect_outlier_report rdf:type prov:Activity .',
    ':inspect_outlier_report sc:isPartOf :data_understanding_phase .',
    ':inspect_outlier_report rdfs:comment \'Data Understanding\' .',
    f':inspect_outlier_report rdfs:comment """{ior_comment}""" .',
    f':inspect_outlier_report prov:startedAtTime "{start_time_ior}"^^xsd:dateTime .',
    f':inspect_outlier_report prov:endedAtTime "{now()}"^^xsd:dateTime .',

    f':inspect_outlier_report prov:qualifiedAssociation :{ior_ass_uuid_executor} .',
    f':{ior_ass_uuid_executor} prov:agent :{inspect_outlier_report_executor} .',
    f':{ior_ass_uuid_executor} rdf:type prov:Association .',
    f':{ior_ass_uuid_executor} prov:hadRole :{code_executor_role} .',

    ':inspect_outlier_report prov:used :outlier_report .',

    ':outlier_decision rdf:type prov:Entity .',
    ':outlier_decision rdfs:comment "Decision: no outliers removed in Data Understanding; consider robust handling in Data Preparation if needed." .',
    ':outlier_decision prov:wasGeneratedBy :inspect_outlier_report .',
]
engine.insert(inspect_outlier_report_activity, prefixes=prefixes)
```

**2f) Risks of Bias**

```python
bias_code_writer = student_a
start_time_2f = now()

# --- normalize column names (no helper function) ---
data_ = data.copy()
data_.columns = [c.strip().replace("\t", "") for c in data_.columns]

TARGET_COL = "Target" if "Target" in data_.columns else None

# Bias-relevant
candidate_cols = [
    "Gender",
    "Age at enrollment",
    "Nacionality",
    "Marital status",
    "Displaced",
    "Educational special needs",
    "Scholarship holder",
    "Debtor",
    "Tuition fees up to date",
]
bias_cols = [c for c in candidate_cols if c in data_.columns]

# Bin age for more stable groups (avoid many unique numeric ages)
if "Age at enrollment" in bias_cols:
    data_["Age at enrollment (binned)"] = pd.cut(
        data_["Age at enrollment"],
        bins=[-np.inf, 17, 21, 25, 30, np.inf],
        labels=["<=17", "18-21", "22-25", "26-30", ">=31"]
    )
    bias_cols.remove("Age at enrollment")
    bias_cols.insert(1, "Age at enrollment (binned)")

# --- 1) Target imbalance (relevant bias risk for modeling/evaluation) ---
if TARGET_COL is not None:
    target_dist = data_[TARGET_COL].value_counts(dropna=False).to_frame("count")
    target_dist["pct"] = (target_dist["count"] / len(data_) * 100).round(2)
else:
    target_dist = None

# --- 2) Underrepresentation / small group risk (simple heuristic) ---
MIN_PCT = 5.0
MIN_N = 50

underrep_rows = []
for col in bias_cols:
    n = len(data_)
    vc = data_[col].value_counts(dropna=True)
    pct = (vc / n * 100)

    flagged = ((pct < MIN_PCT) | (vc < MIN_N)).sum()
    underrep_rows.append({
        "column": col,
        "n_unique_non_missing": int(len(vc)),
        "n_groups_flagged_small": int(flagged),
    })

underrep_df = pd.DataFrame(underrep_rows).sort_values(
    ["n_groups_flagged_small", "n_unique_non_missing"],
    ascending=False
)

# --- 3) Outcome differences across groups (only a few key columns, no Nacionality tables) ---
KEY_COLS = [
    "Gender",
    "Age at enrollment (binned)",
    "Debtor",
    "Tuition fees up to date",
    "Scholarship holder",
    "Educational special needs",
]
KEY_COLS = [c for c in KEY_COLS if c in bias_cols]

disparity_rows = []
if TARGET_COL is not None:
    for col in KEY_COLS:
        ct = pd.crosstab(data_[col], data_[TARGET_COL], dropna=False)

        # group sizes (exclude extremely tiny groups from disparity calc)
        group_size = ct.sum(axis=1)
        ct_valid = ct[group_size >= 30]  # stability threshold
        if ct_valid.shape[0] < 2:
            continue

        ct_pct = (ct_valid.div(ct_valid.sum(axis=1), axis=0) * 100)

        # compute range (max-min) for each class if present
        row = {"column": col, "n_groups_used": int(ct_valid.shape[0])}
        for cls in ["Dropout", "Enrolled", "Graduate"]:
            if cls in ct_pct.columns:
                row[f"{cls}_range_pp"] = float((ct_pct[cls].max() - ct_pct[cls].min()).round(2))
        disparity_rows.append(row)

disparity_df = pd.DataFrame(disparity_rows).sort_values(
    [c for c in ["Dropout_range_pp", "Graduate_range_pp"] if c in pd.DataFrame(disparity_rows).columns],
    ascending=False
) if len(disparity_rows) > 0 else pd.DataFrame()

# =========================
# OUTPUT (only what you need)
# =========================
print("2f Risks of Bias — Evidence tables (minimal)")

if target_dist is not None:
    print("\n(1) Target distribution (class imbalance):")
    display(target_dist)

print("\n(2) Underrepresentation overview (groups flagged if <5% or <50 rows):")
display(underrep_df)

if TARGET_COL is not None and not disparity_df.empty:
    print("\n(3) Outcome differences across groups (range in percentage points; computed only on groups with n>=30):")
    display(disparity_df)

# =========================
# TEXTUAL ANSWER (core of 2f)
# =========================
minority_cols = underrep_df.loc[underrep_df["n_groups_flagged_small"] > 0, "column"].tolist()

notable_disp = []
if TARGET_COL is not None and not disparity_df.empty and "Dropout_range_pp" in disparity_df.columns:
    notable_disp = disparity_df.loc[disparity_df["Dropout_range_pp"] >= 15, "column"].tolist()

du2f_comment = f"""
2f Risks of Bias:

Potential risks / bias types indicated by the data exploration:
- Target imbalance: {("present (see table)" if TARGET_COL is not None else "Target not available")}. This can bias training and evaluation (micro vs macro metrics, stratified splits, resampling).
- Underrepresented groups: {minority_cols if minority_cols else "none flagged by heuristic"}.
  Small groups (especially for high-cardinality attributes like Nacionality) can lead to unstable estimates and misleading fairness conclusions.
- Group outcome differences (if Target exists): columns with notable dropout-rate spread (>=15pp heuristic): {notable_disp if notable_disp else "none flagged"}.
  Such differences can come from bias, confounding, or real structural differences; they must be interpreted with domain context.
- Proxy risk / socioeconomic bias: features such as Debtor / Tuition fees up to date / Scholarship holder can encode socioeconomic status and may cause disparate impact.
- Selection & measurement bias (needs confirmation): dataset scope (which programs/years), and how "Dropout/Graduate/Enrolled" are defined can systematically skew labels.

Questions for an external expert / data owner to assess bias and data quality:
1) Data scope & representativeness:
   - Which institution(s), programs, cohorts/years are included? Any known coverage gaps (e.g., part-time, international, specific faculties)?
2) Label validity:
   - How exactly are 'Dropout', 'Enrolled', 'Graduate' defined? Are 'Enrolled' cases censored (still studying), and how is that handled?
3) Temporal order / leakage:
   - When are variables like tuition fee status, debtor status, scholarship recorded relative to the outcome? Could they be measured after (or because of) dropout?
4) Semantics of codes:
   - What do the coded categories mean (e.g., Nacionality IDs, Marital status codes)? Any known recoding changes over time?
5) Policy/operational confounding:
   - Are there policies/interventions that differ by subgroup (e.g., scholarship allocation rules, program-specific dropout handling)?
6) Intended use & constraints:
   - For the intended decision, which attributes are legally/ethically allowed? Are protected attributes excluded, used only for auditing, or required for fairness evaluation?

Interpretation
The analysis indicates several potential bias risks in the dataset. 
The target variable is moderately imbalanced, which may affect model training and evaluation if not handled appropriately. 
Some attributes, particularly Nacionality, Marital status, Age at enrollment (binned) and Educational special needs, 
contain underrepresented groups with very small sample sizes, limiting the reliability of group-based comparisons. 
In addition, noticeable differences in outcome rates are observed for variables such as Tuition fees up to date, Debtor, Scholarship holder, Age at enrollment (binned) and Gender. 
These differences should be interpreted with caution, as they may result from confounding factors, socioeconomic proxy effects, or temporal proximity to the outcome rather than direct bias.
""".strip()

print("\n" + du2f_comment)

end_time_2f = now()

# =========================
# PROV logging
# =========================
du2f_ass_uuid_exec = str(uuid.uuid4())
engine.insert([
    f':du_risks_of_bias prov:qualifiedAssociation :{du2f_ass_uuid_exec} .',
    f':{du2f_ass_uuid_exec} prov:agent :{executed_by} .',
    f':{du2f_ass_uuid_exec} rdf:type prov:Association .',
    f':{du2f_ass_uuid_exec} prov:hadRole :{code_executor_role} .'
], prefixes=prefixes)

du2f_ass_uuid_writer = str(uuid.uuid4())
engine.insert([
    ':du_risks_of_bias rdf:type prov:Activity .',
    ':du_risks_of_bias sc:isPartOf :data_understanding_phase .',
    ':du_risks_of_bias rdfs:comment "Data Understanding 2f: Risks of Bias" .',
    f':du_risks_of_bias rdfs:comment """{du2f_comment}""" .',
    f':du_risks_of_bias prov:startedAtTime "{start_time_2f}"^^xsd:dateTime .',
    f':du_risks_of_bias prov:endedAtTime "{end_time_2f}"^^xsd:dateTime .',

    f':du_risks_of_bias prov:qualifiedAssociation :{du2f_ass_uuid_writer} .',
    f':{du2f_ass_uuid_writer} prov:agent :{bias_code_writer} .',
    f':{du2f_ass_uuid_writer} rdf:type prov:Association .',
    f':{du2f_ass_uuid_writer} prov:hadRole :{code_writer_role} .',

    ':du_risks_of_bias prov:used :data .',
    ':du2f_bias_report rdf:type prov:Entity .',
    ':du2f_bias_report prov:wasGeneratedBy :du_risks_of_bias .'
], prefixes=prefixes)
```

**3a) Data preprocessing**

```python
import pandas as pd
import numpy as np
import json

apply_preprocessing_code_writer = student_b  

def _ttl_multiline(s: str) -> str:
    return s.replace("\\", "\\\\").replace('"""', '\\"""')

def apply_preprocessing(
    data: pd.DataFrame,
    decision: str,
    *,
    # 2a --> tab/whitespace in column names
    fix_column_whitespace: bool = True,

    # safe defaults
    drop_duplicates: bool = True,

    # 2g: --> coded categories as categorical labels
    categorical_coded_cols=(
        "Marital status", "Course", "Previous qualification",
        "Nacionality", "Mother's qualification", "Father's qualification",
        "Mother's occupation", "Father's occupation",
        "Displaced", "Educational special needs", "Debtor",
        "Tuition fees up to date", "Scholarship holder", "Gender",
        "Application mode", "Application order", "Daytime/evening attendance"
    ),

    # 2g + DU 2e/2f --> high-cardinality handling (Nationality)
    reduce_high_cardinality: bool = True,
    high_cardinality_col: str = "Nacionality",
    min_group_n: int = 50,
    min_group_pct: float = 5.0,

    # DU 2e/2f --> stable age groups for auditing
    add_age_bins: bool = True,

    # outliers --> do not remove rows; robust handling if needed
    outlier_strategy: str = "winsorize",  # "leave" | "winsorize"
    outlier_cols=(
        "Age at enrollment",
        "Admission grade",
        "Previous qualification (grade)",
        "Curricular units 1st sem (grade)",
        "Curricular units 2nd sem (grade)",
        "Unemployment rate",
        "Inflation rate",
        "GDP",
    ),
    winsor_q_low: float = 0.01,
    winsor_q_high: float = 0.99,
) -> tuple[pd.DataFrame, dict]:

    df = data.copy(deep=True)

    log = {"params": {}, "changes": {}}
    log["changes"]["shape_before"] = tuple(df.shape)

    # 1) 2a: Clean column names (strip whitespace + remove tabs)
    if fix_column_whitespace:
        before_cols = list(df.columns)
        df.columns = [c.strip().replace("\t", "") for c in df.columns]
        renamed = [(b, a) for b, a in zip(before_cols, df.columns) if b != a]
        log["changes"]["columns_cleaned_examples"] = renamed[:8]

    # 2) Drop duplicates 
    if drop_duplicates:
        n_before = len(df)
        df = df.drop_duplicates()
        log["changes"]["duplicates_removed"] = int(n_before - len(df))

    # 3) Missing values check 
    n_missing_total = int(df.isna().sum().sum())
    log["changes"]["missing_total"] = n_missing_total
    log["changes"]["missing_action"] = "none_detected" if n_missing_total == 0 else "present_not_handled_in_3a"

    # 4) 2g --> coded categoricals to category
    casted = []
    for c in categorical_coded_cols:
        if c in df.columns:
            df[c] = df[c].astype("category")
            casted.append(c)
    log["changes"]["categorical_casted"] = casted

    # 5) 2g/2e/2f --> Reduce high-cardinality for Nationality
    if reduce_high_cardinality and high_cardinality_col in df.columns:
        vc = df[high_cardinality_col].value_counts(dropna=True)
        n = len(df)
        pct = (vc / n * 100)
        rare_levels = set(vc[(vc < min_group_n) | (pct < min_group_pct)].index.tolist())

        new_col = f"{high_cardinality_col}_reduced"
        df[new_col] = df[high_cardinality_col].apply(
            lambda x: "Other" if x in rare_levels else x
        ).astype("category")

        log["changes"]["high_cardinality_reduction"] = {
            "column": high_cardinality_col,
            "new_column": new_col,
            "n_rare_levels_grouped": int(len(rare_levels)),
            "thresholds": {"min_group_n": min_group_n, "min_group_pct": min_group_pct},
        }

    # 6) 2e/2f --> Add age bins
    if add_age_bins and "Age at enrollment" in df.columns:
        age_num = pd.to_numeric(df["Age at enrollment"], errors="coerce")
        df["Age at enrollment (binned)"] = pd.cut(
            age_num,
            bins=[-np.inf, 17, 21, 25, 30, np.inf],
            labels=["<=17", "18-21", "22-25", "26-30", ">=31"]
        ).astype("category")
        log["changes"]["age_binning"] = True

    # 7) outliers --> robust handling WITHOUT row removal (winsorization)
    wins_changes = {}
    if outlier_strategy == "winsorize":
        for col in outlier_cols:
            if col not in df.columns:
                continue
            s = pd.to_numeric(df[col], errors="coerce")
            if s.notna().sum() == 0:
                continue
            lo, hi = s.quantile(winsor_q_low), s.quantile(winsor_q_high)
            clipped = s.clip(lo, hi)
            changed = int((s != clipped).sum())
            df[col] = clipped
            wins_changes[col] = {"low": float(lo), "high": float(hi), "n_cells_changed": changed}
        log["changes"]["winsorization"] = wins_changes
    else:
        log["changes"]["winsorization"] = "not_applied"

    log["changes"]["shape_after"] = tuple(df.shape)

    log["params"] = {
        "decision": decision,
        "fix_column_whitespace": fix_column_whitespace,
        "drop_duplicates": drop_duplicates,
        "categorical_coded_cols": list(categorical_coded_cols),
        "reduce_high_cardinality": reduce_high_cardinality,
        "high_cardinality_col": high_cardinality_col,
        "min_group_n": min_group_n,
        "min_group_pct": min_group_pct,
        "add_age_bins": add_age_bins,
        "outlier_strategy": outlier_strategy,
        "outlier_cols": list(outlier_cols),
        "winsor_q_low": winsor_q_low,
        "winsor_q_high": winsor_q_high,
    }

    return df, log


# --------------------
# Execute preprocessing (3a)
# --------------------
start_time_dp_a = now()

decision_summary = (
    "DU decisions: clean column-name tab issue (DU 2a); treat coded categories as categorical (DU 2g); "
    "reduce rare Nacionality groups using <5% or <50 heuristic (DU 2e/2f/2g); "
    "do not remove outliers; use robust handling if needed (outlier decision)."
)

cleaned_data, dp_a_log = apply_preprocessing(
    data,
    decision=decision_summary,
    fix_column_whitespace=True,
    drop_duplicates=True,
    reduce_high_cardinality=True,
    high_cardinality_col="Nacionality",
    min_group_n=50,
    min_group_pct=5.0,
    add_age_bins=True,
    outlier_strategy="winsorize",
    winsor_q_low=0.01,
    winsor_q_high=0.99,
)

end_time_dp_a = now()

display(cleaned_data.head())


#############################################
# Documentation (2b)
#############################################

min_group_n = dp_a_log["params"]["min_group_n"]
min_group_pct = dp_a_log["params"]["min_group_pct"]
winsor_q_low = dp_a_log["params"]["winsor_q_low"]
winsor_q_high = dp_a_log["params"]["winsor_q_high"]

cols_cleaned_examples = dp_a_log["changes"].get("columns_cleaned_examples", [])
n_cat_casted = len(dp_a_log["changes"].get("categorical_casted", []))
nat_red = dp_a_log["changes"].get("high_cardinality_reduction", {})
wins_dict = dp_a_log["changes"].get("winsorization", {})
wins_cols = list(wins_dict.keys()) if isinstance(wins_dict, dict) else []
age_binned = bool(dp_a_log["changes"].get("age_binning", False))

derived_cols = []
if age_binned:
    derived_cols.append("Age at enrollment (binned)")
if nat_red:
    derived_cols.append(nat_red.get("new_column", "Nacionality_reduced"))

dp_a_comment = f"""
3a Data Preparation (execution + reproducibility log):
- Input dataset: rows={dp_a_log["changes"]["shape_before"][0]}, cols={dp_a_log["changes"]["shape_before"][1]}
- Column names cleaned (whitespace/tab). Examples: {cols_cleaned_examples}
- Duplicates removed: {dp_a_log["changes"].get("duplicates_removed", 0)}
- Missing values: total_missing={dp_a_log["changes"].get("missing_total")}, action={dp_a_log["changes"].get("missing_action")}
- Categorical-coded columns cast to category: {n_cat_casted}
- Nacionality reduction: {'applied' if nat_red else 'not applied'} (min_n={min_group_n}, min_pct={min_group_pct}%)
- Age binning: {'applied' if age_binned else 'not applied'}
- Outliers: winsorize={'applied' if wins_cols else 'not applied'} on {len(wins_cols)} cols (q={winsor_q_low}-{winsor_q_high}); no rows removed
- Derived columns added: {derived_cols if derived_cols else 'none'}
- Output dataset: rows={dp_a_log["changes"]["shape_after"][0]}, cols={dp_a_log["changes"]["shape_after"][1]}
""".strip()

print(dp_a_comment)


# --------------------
# KG documentation (PROV)
# --------------------

# 1) Executor association
dp_a_ass_uuid_executor = "d6b5d0d5-3d6c-4c6f-8fa1-0a1e5a7a9c10"  # generate once, keep stable
dp_a_executor_triples = [
    f':dp_apply_preprocessing prov:qualifiedAssociation :{dp_a_ass_uuid_executor} .',
    f':{dp_a_ass_uuid_executor} prov:agent :{executed_by} .',
    f':{dp_a_ass_uuid_executor} rdf:type prov:Association .',
    f':{dp_a_ass_uuid_executor} prov:hadRole :{code_executor_role} .',
]
engine.insert(dp_a_executor_triples, prefixes=prefixes)

# 2) Activity + writer association + lineage
dp_a_ass_uuid_writer = "2d3a5d2e-7b0a-4b88-9d2a-6b2de2e7d1aa"  
dp_a_comment_ttl = _ttl_multiline(dp_a_comment)

dp_a_activity_triples = [
    ':dp_apply_preprocessing rdf:type prov:Activity .',
    ':dp_apply_preprocessing sc:isPartOf :data_preparation_phase .',
    ':dp_apply_preprocessing rdfs:comment "Data Preparation (3a)" .',
    f':dp_apply_preprocessing rdfs:comment """{dp_a_comment_ttl}""" .',
    f':dp_apply_preprocessing prov:startedAtTime "{start_time_dp_a}"^^xsd:dateTime .',
    f':dp_apply_preprocessing prov:endedAtTime "{end_time_dp_a}"^^xsd:dateTime .',

    f':dp_apply_preprocessing prov:qualifiedAssociation :{dp_a_ass_uuid_writer} .',
    f':{dp_a_ass_uuid_writer} prov:agent :{apply_preprocessing_code_writer} .',
    f':{dp_a_ass_uuid_writer} rdf:type prov:Association .',
    f':{dp_a_ass_uuid_writer} prov:hadRole :{code_writer_role} .',

    # Link to DU outputs / decisions
    ':dp_apply_preprocessing prov:used :data .',
    ':dp_apply_preprocessing prov:used :outlier_decision .',
    ':dp_apply_preprocessing prov:used :du_actions_required .',

    # Output dataset
    ':cleaned_data rdf:type prov:Entity .',
    ':cleaned_data prov:wasGeneratedBy :dp_apply_preprocessing .',
    ':cleaned_data prov:wasDerivedFrom :data .',
]
engine.insert(dp_a_activity_triples, prefixes=prefixes)
```

**3b) Other pre-processing steps considered**

1) Missing values: even though imputation was considered to deal with missing values, since there were no missing values in the dataset there was no need of handling missing values

2) Row-wise outlier removal: Removing observations with extreme values based on z-scores was considered but later not required since the outlier analysis showed that detected outliers were plausible rare cases. This led to a decision of not removing rows. Instead robust handling (winsorization) was which limits the influence of extreme values while preserving observations.

3) Feature scaling/ normalization: Standardization or normalization of numeric variables was considered for feature scaling but we will deal with it in the model phase.

4) Attribute removal due to correlation: since 2b) showed that removal of correlated attributes was necessary.

5) Binning of numeric variables: Binning continuous variables such as grades were avoided due to ethical auditing and group stability.

**3c) Potential derived attributes**

Some derived attributes have been dealt in 3a) like binig age and nacionality reuduction. 
There are other options to achieve derived attributes.
- Age-based derived attributes: 2d showed a right-skewed age distribution and 2e and 2f identified age as an ethically sensitive attribute, where stable groups are needed for bias auditing and interpretability. This led to a conclusion to bin age into stable groups. This improves interpretability, reduces noise from individual ages and enables fair group-level analysis without exposing fine-grained sensitive information.
- Aggregation of academic performance indicators: 2d visualizations showed that academic performance variables differ strongly across target classes and carry predictive signal. This could lead to creating composite indicators such as total approved curricular units, average grade across semesters or pass rate (approved / evaluated units). But this could lead to data leakage depending on when grades are recorded relative to dropout/graduate outcomes. Therefore it is wise to use it during modelling.
- Socioeconomic proxy indicators: We could combine variables such as Debtor, Tuition fees up to date, and Scholarship holder into a single socioeconomic risk indicator as 2f shows that these are strongly associated with outcome differences. But since ggregation could obscure important distinctions we decide on not applying it.
- Interaction features: d2 suggested that performance signals may vary across groups. But we don't use it since Iinteraction effects are model-specific. And with increasing feature dimensionality the risk of overfitting also could happen. 
- Temporal or progression-based attributes: even though there are multiple semester-based academic indicators, we avoid it since it requires clear temporal alignment with the target label and also risks of post-outcome leakage if not carefully defined. 

```python
import pandas as pd
import numpy as np
import json

derive_features_code_writer = student_b  

def _ttl_multiline(s: str) -> str:
    return s.replace("\\", "\\\\").replace('"""', '\\"""')

def derive_attributes(
    df: pd.DataFrame,
    *,
    # Option 1: Age bins
    add_age_bins: bool = True,
    age_col: str = "Age at enrollment",
    age_bins = [-np.inf, 17, 21, 25, 30, np.inf],
    age_labels = ["<=17", "18-21", "22-25", "26-30", ">=31"],

    # Option 3: Academic aggregates
    add_academic_aggregates: bool = False,
    # You can adjust these lists to match available columns
    approved_cols = ("Curricular units 1st sem (approved)", "Curricular units 2nd sem (approved)"),
    grade_cols = ("Curricular units 1st sem (grade)", "Curricular units 2nd sem (grade)"),
    evaluated_cols = ("Curricular units 1st sem (evaluations)", "Curricular units 2nd sem (evaluations)"),

    # Option 4: Simple progression deltas (2nd - 1st)
    add_progression_deltas: bool = False,

    # Option 5: Socioeconomic composite (ethically sensitive -> usually keep for auditing)
    add_ses_index: bool = False,
    ses_cols = ("Debtor", "Tuition fees up to date", "Scholarship holder"),

    # Option 6: Interactions (typically model-specific)
    add_interactions: bool = False,
    interaction_pairs = (("Age at enrollment (binned)", "Course"),),

) -> tuple[pd.DataFrame, dict]:
    """
    3(c) Derived attributes: implement OPTIONS and return (df_derived, log_dict).
    The log documents what was applied and which columns were created.
    """
    out = df.copy(deep=True)

    created = []
    log = {"params": {}, "created_columns": [], "notes": []}

    # helper: check col existence
    def _has_all(cols):
        return all(c in out.columns for c in cols)

    # 1) Age bins
    if add_age_bins and age_col in out.columns:
        age_num = pd.to_numeric(out[age_col], errors="coerce")
        out["Age at enrollment (binned)"] = pd.cut(age_num, bins=age_bins, labels=age_labels).astype("category")
        created.append("Age at enrollment (binned)")
        
    log["created_columns"] = created
    log["params"] = {
        "add_age_bins": add_age_bins,
        "reduce_high_cardinality": reduce_high_cardinality,
        "min_group_n": min_group_n,
        "min_group_pct": min_group_pct,
        "add_academic_aggregates": add_academic_aggregates,
        "add_progression_deltas": add_progression_deltas,
        "add_ses_index": add_ses_index,
        "add_interactions": add_interactions,
    }
    return out, log
```

```python
# This cell includes cleaning functions

from datetime import datetime

def latex_escape(text: str | None) -> str:
    if text is None: return ""
    text = str(text)
    text = text.replace("\\", r"\textbackslash{}")
    pairs = [
        ("&", r"\&"), ("%", r"\%"), ("$", r"\$"), ("#", r"\#"), 
        ("_", r"\_"), ("{", r"\{"), ("}", r"\}"), 
        ("~", r"\textasciitilde{}"), ("^", r"\textasciicircum{}")
    ]
    for k, v in pairs:
        text = text.replace(k, v)
    return text

def clean_rdf(x) -> str:
    if hasattr(x, "toPython"): return str(x.toPython())
    if x is None: return ""
    s = str(x).strip()
    s = s.strip('"').strip("'")
    s = s.strip()
    if "^^" in s:
        s = s.split("^^")[0].strip('"')
        
    return s

def fmt_iso(ts: str) -> str:
    if not ts: return ""
    try:
        clean_ts = ts.split("^^")[0].strip('"')
        clean_ts = clean_ts.replace("Z", "+00:00") if clean_ts.endswith("Z") else clean_ts
        return datetime.fromisoformat(clean_ts).strftime("%Y-%m-%d %H:%M:%S")
    except:
        return latex_escape(str(ts))
```

```python
# =========================
# INTERIM REPORT GENERATOR (Sections 1–3) 
# =========================

from pathlib import Path

OUT_TEX = "interim_report.tex"   # <- output filename

# ---------- helpers ----------
def clean_rdf(x):
    if x is None:
        return ""
    return str(x).strip()

def latex_escape(s: str) -> str:
    if s is None:
        return ""
    s = str(s)
    repl = {
        '\\': r'\textbackslash{}',
        '&': r'\&',
        '%': r'\%',
        '$': r'\$',
        '#': r'\#',
        '_': r'\_',
        '{': r'\{',
        '}': r'\}',
        '~': r'\textasciitilde{}',
        '^': r'\textasciicircum{}',
    }
    for k, v in repl.items():
        s = s.replace(k, v)
    return s

def get_one(q: str, var: str):
    res = engine.query(q)
    if res is None or res.empty:  # type:ignore
        return ""
    return clean_rdf(res.iloc[0].get(var, ""))  # type:ignore

def get_comment(entity: str) -> str:
    q = f"""
{prefix_header}
SELECT ?c WHERE {{
  :{entity} rdfs:comment ?c .
}} LIMIT 1
"""
    return latex_escape(get_one(q, "c"))

def list_phase_comments(phase_id: str):
    """
    Robust: returns list of (label, comment) for ALL entities in a phase.
    No GROUP BY / SAMPLE, so nothing gets swallowed.
    Dedupe identical (label, comment) pairs.
    """
    q = f"""
{prefix_header}
SELECT ?label ?c WHERE {{
  ?e prov:wasGeneratedBy ?a .
  ?a sc:isPartOf :{phase_id} .
  OPTIONAL {{ ?e rdfs:label ?label . }}
  ?e rdfs:comment ?c .
}}
ORDER BY ?label ?c
"""
    res = engine.query(q)
    items = []
    if res is None or res.empty:  # type:ignore
        return items

    seen = set()
    for _, r in res.iterrows():  # type:ignore
        lbl_raw = clean_rdf(r.get("label", ""))
        com_raw = clean_rdf(r.get("c", ""))
        if not com_raw:
            continue

        lbl = latex_escape(lbl_raw)
        com = latex_escape(com_raw)

        key = (lbl, com)
        if key in seen:
            continue
        seen.add(key)

        items.append((lbl, com))
    return items


# ---------- Author block ----------
author_query = f"""
{prefix_header}
PREFIX iao: <http://purl.obolibrary.org/obo/>

SELECT DISTINCT ?uri ?given ?family ?matr WHERE {{
  VALUES ?uri {{ :{student_a} :{student_b} }}
  ?uri a foaf:Person .
  ?uri foaf:givenName ?given .
  ?uri foaf:familyName ?family .
  ?uri iao:IAO_0000219 ?matr .
}}
"""
res_authors = engine.query(author_query)
author_block_latex = ""

if res_authors is not None and not res_authors.empty:  # type:ignore
    for _, row in res_authors.iterrows():  # type:ignore
        uri_str = str(row["uri"])
        given = latex_escape(clean_rdf(row["given"]))
        family = latex_escape(clean_rdf(row["family"]))
        matr = latex_escape(clean_rdf(row["matr"]))

        if student_a in uri_str:
            responsibility = "Student A"
        elif student_b in uri_str:
            responsibility = "Student B"
        else:
            responsibility = "Student"

        author_block_latex += rf"""
\author{{{given} {family}}}
\authornote{{{responsibility}, Matr.Nr.: {matr}}}
\affiliation{{
  \institution{{TU Wien}}
  \country{{Austria}}
}}
"""

# ---------- Business Understanding 1a–1f ----------
bu_1a = get_comment("bu_data_source_and_scenario")
bu_1b = get_comment("bu_business_objectives")
bu_1c = get_comment("bu_business_success_criteria")
bu_1d = get_comment("bu_data_mining_goals")
bu_1e = get_comment("bu_data_mining_success_criteria")
bu_1f = get_comment("bu_ai_risk_aspects")

bu_section_latex = rf"""
\section{{Business Understanding}}

\subsection{{Data Source and Scenario}}
{bu_1a}

\subsection{{Business Objectives}}
{bu_1b}

\subsection{{Business Success Criteria}}
{bu_1c}

\subsection{{Data Mining Goals}}
{bu_1d}

\subsection{{Data Mining Success Criteria}}
{bu_1e}

\subsection{{AI Risk Aspects}}
{bu_1f}
"""

# ---------- Data Understanding (description + features table + ALL comments) ----------
du_desc_query = f"""
{prefix_header}
SELECT (SAMPLE(?desc) AS ?desc) WHERE {{
  OPTIONAL {{ :raw_data sc:description ?desc . }}
  OPTIONAL {{ :data sc:description ?desc . }}
}} LIMIT 1
"""
du_description = latex_escape(get_one(du_desc_query, "desc"))

# Feature metadata (optional; if you didn't log Croissant fields, this table will show a fallback row)
du_features_query = f"""
{prefix_header}
SELECT ?name (SAMPLE(?dtypeRaw) as ?dtype) (SAMPLE(?descRaw) as ?desc) WHERE {{
  ?ds cr:recordSet ?rs .
  ?rs cr:field ?field .
  ?field sc:name ?name .
  OPTIONAL {{ ?field sc:description ?descRaw . }}
  OPTIONAL {{ ?field cr:dataType ?dtypeRaw . }}
  FILTER (?ds = :raw_data || ?ds = :data)
}}
GROUP BY ?name
ORDER BY ?name
"""
res_du_feat = engine.query(du_features_query)
du_rows = []

if res_du_feat is not None and not res_du_feat.empty:  # type:ignore
    for _, f in res_du_feat.iterrows():  # type:ignore
        dtype_raw = clean_rdf(f.get("dtype", ""))
        if "#" in dtype_raw:
            dtype = dtype_raw.split("#")[-1]
        elif "/" in dtype_raw:
            dtype = dtype_raw.split("/")[-1]
        else:
            dtype = dtype_raw

        desc = clean_rdf(f.get("desc", ""))
        du_rows.append(
            f"{latex_escape(clean_rdf(f['name']))} & {latex_escape(dtype)} & {latex_escape(desc)} \\\\"
        )
else:
    du_rows.append(r"\multicolumn{3}{l}{No feature metadata logged (Croissant fields missing).} \\")

du_table_rows = "\n".join(du_rows)

# Pull ALL DU comments from the phase (this is the key fix)
du_items = list_phase_comments("data_understanding_phase")
if not du_items:
    du_items = [("", "No detailed Data Understanding comments were found in the KG.")]

du_items_latex = "\n".join(
    [rf"\item \textbf{{{lbl}}}: {com}" if lbl else rf"\item {com}" for (lbl, com) in du_items]
)

du_section_latex = rf"""
\section{{Data Understanding}}

\subsection{{Dataset Description}}
{du_description}

\subsection{{Raw Data Features}}
\begin{{table}}[H]
\caption{{Raw Data Features}}
\centering
\begin{{tabular}}{{lll}}
\toprule
Feature Name & Data Type & Description \\
\midrule
{du_table_rows}
\bottomrule
\end{{tabular}}
\end{{table}}

\subsection{{Findings and Notes (from KG comments)}}
\begin{{itemize}}
{du_items_latex}
\end{{itemize}}
"""

# ---------- Data Preparation (ALL comments) ----------
dp_items = list_phase_comments("data_preparation_phase")
if not dp_items:
    dp_items = [("", "No Data Preparation comments were found in the KG (did you execute the DP logging cells?).")]

dp_items_latex = "\n".join(
    [rf"\item \textbf{{{lbl}}}: {com}" if lbl else rf"\item {com}" for (lbl, com) in dp_items]
)

dp_section_latex = rf"""
\section{{Data Preparation}}
\subsection{{Steps, Decisions, and Rationale (from KG comments)}}
\begin{{itemize}}
{dp_items_latex}
\end{{itemize}}
"""

# ---------- Compose full ACM paper (Interim: only Sections 1–3) ----------
paper_tex = rf"""
\documentclass[sigconf]{{acmart}}

\AtBeginDocument{{ \providecommand\BibTeX{{ Bib\TeX }} }}

\usepackage{{booktabs}}
\usepackage{{float}}

\title{{BI2025 Experiment Report - Group {group_id}}}

{author_block_latex}

\begin{{document}}
\maketitle

\begin{{abstract}}
This report documents the machine learning experiment for Group {group_id}, following the CRISP-DM process model.
\end{{abstract}}

\ccsdesc[500]{{Computing methodologies~Machine learning}}
\keywords{{CRISP-DM, Provenance, Knowledge Graph, Machine Learning}}

{bu_section_latex}
{du_section_latex}
{dp_section_latex}

\section*{{Roles and Responsibilities}}
Student A and Student B contributed to different phases of the CRISP-DM process.
Responsibilities were clearly assigned and are documented in the provenance metadata via explicit role annotations
(code writer / code executor).

\end{{document}}
"""

# Write file
Path(OUT_TEX).write_text(paper_tex, encoding="utf-8")

print(f"✅ Wrote LaTeX interim report to: {OUT_TEX}")
print("Sanity checks:")
print("  BU lengths:", len(bu_1c), len(bu_1d), len(bu_1e), len(bu_1f))
print("  DU items:", len(du_items))
print("  DP items:", len(dp_items))
print("  NOTE: In Overleaf, ensure you uploaded the NEW interim_report.tex and recompiled.")
```
